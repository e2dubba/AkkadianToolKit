```python
def cuneiform_sep(cell):
    cunei = []
    others = []
    for char in cell:
        if ord(char) > 73727 and ord(char) < 75076:
            cunei.append(char)
        else:
            others.append(char)
    return ''.join(cunei), ''.join(others) 
```

```python
def find_unicode(labat_number, df):
    row = df[df['≈†L/HA'] == labat_number]
    return cuneiform_sep(row['Unicode Code point'].values[0])
```

```python
def cunei_recursion(first_span, second_span, span_list, example, cuneiform, latin):
    latin += example[second_span[0]:second_span[1]]
    cuneiform += example[first_span[1]:second_span[0]]
    for span in span_list:
        latin, cuneiform = cunei_recursion(second_span, span_list[0], span_list[1:], example, cuneiform, latin)
        return latin, cuneiform 
    return latin, cuneiform
```

```python
values = {'examples in New Assyrian orthography',
          'phonetic, Ass. old', 'phonetic, old', 'phonetic, Ass. new',
          'phonetic, new',
          'phonetic, mostly new',
          'phonetic, mostly old',
          'phonetic, bab.',
          'examples in New Assyrian orthography',
          'phonetic, Ass.',
          'phonetic, Ass.new',
          'new',
          'phonetic, Bab. new',
          'phonetic',
         }
```

```python
new_syllabar = {}
for key in syllabar.keys():
    phonetic = set()
    for ky in syllabar[key].keys():
        if ky != 'values':
            new_syllabar.setdefault(key, {})[ky] = syllabar[key][ky]
    sign_meanings = syllabar[key]['values']
    for grandkid in sign_meanings.keys():
        if grandkid in values:
            for phono_value in sign_meanings[grandkid]:
                phonetic.add(phono_value)
        else:
            new_syllabar[key].setdefault('values', {})[grandkid] = sign_meanings[grandkid]
    new_syllabar.setdefault(key, {}).setdefault('values', {})['phonetic'] = list(phonetic)
        
        
```
